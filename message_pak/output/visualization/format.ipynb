```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
from pathlib import Path
```

```python
def format(rep_out_df, filter_reg, scen_name, rename_reg, rename_model):
    rep_out_df = rep_out_df[rep_out_df["Region"] == filter_reg]
    rep_out_df["Scenario"] = scen_name
    rep_out_df["Region"] = rename_reg
    rep_out_df["Model"] = rename_model
    rep_out_df = rep_out_df.drop_duplicates()

    rep_out_df.loc[rep_out_df["Unit"] == "US$2010/kW", "Unit"] = "USD_2010/kW"
    rep_out_df.loc[rep_out_df["Unit"] == "billion US$2010/yr", "Unit"] = "billion USD_2010/yr"
    rep_out_df.loc[rep_out_df["Unit"] == "US$2010/kW/yr", "Unit"] = "USD_2010/kW/yr"

    rep_out_df.to_excel(f"D:/COMMITTED/Models/Model-Manuscript/NEST-Pakistan/message_pak/output/formatted/{scen_name}_NDC-C.xlsx", index=False)
```

```python
output_dir = Path("D:/COMMITTED/Models/Model-Manuscript/NEST-Pakistan/message_pak/output")
file_path = list(output_dir.glob("MESSAGEix-Pakistan_ndc-c*.xlsx"))[0]

sn = file_path.stem.replace("MESSAGEix-Pakistan 1_", "").split('_')[0]

print(f"Processing: {sn}")
scen_rep = pd.read_excel(file_path)
format(scen_rep, "R12_PAK", sn, "Pakistan", "MESSAGEix-Pakistan 1")
```
