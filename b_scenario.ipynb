```python
F1.columns
```

```python
# Long term
F2 = pd.melt(
    F1,
    id_vars=["Region","Metric","Marker"],
    value_vars=[
        2025,
        2030,
        2035,
        2040,
        2045,
        2050,
        2060,
        2070,     
    ],
)

for Reg in F2["Region"].unique():
    Fn = F2.loc[(F2["Region"] == Reg)]
    g = sns.FacetGrid(
        Fn,
        col="Metric",
        hue="Marker",
        palette=cp,
        sharex=False,
        sharey=False,
        col_wrap=7,
        # subplot_kws={'xlim': (2020, 2100)},
    )
    g.map(sns.lineplot, "variable", "value")
    g.add_legend()
    g.set_xticklabels(rotation=90)
    g.fig.suptitle(Reg)
    g.tight_layout()
    plt.show()
    g.savefig(r"PDF/MultiVariable_Long_term_{}.pdf".format(Reg), bbox_inches='tight')
```
